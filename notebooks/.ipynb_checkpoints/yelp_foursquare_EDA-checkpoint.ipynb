# Yelp

Send a request to Yelp with a small radius (1000m) for all the bike stations in your city of choice. 

```python
with open(r'C:\Users\akobe\OneDrive\Documents\Keys\yelpKey.txt') as content:
    yelpKey = content.read()
```

```python
#getting all the URLs 
allUrlYelp = []
for i in range(102):
    lat = (locValues[i][0])
    lon = (locValues[i][1])
    yelpUrl = "https://api.yelp.com/v3/businesses/search?latitude="+str(lat)+"&longitude="+str(lon)+"&radius=1000&categories=bars%26categories%3Dlodging%26categories%3Dhistoricsite&sort_by=best_match&limit=20"
    allUrlYelp.append(yelpUrl)
```

```python
#accessing APIs, getting jsons
headers = {"accept": "application/json"}
headers['Authorization'] = yelpKey
yelpData = []

for j in range(102):
    yelpResponse = requests.get(allUrlYelp[j], headers=headers)
    yelpData.append(yelpResponse.json())
```

```python
#checking that data loaded 
#yelpData
```

```python
#look at data in a nice way 
from IPython.display import JSON
JSON(yelpData)
```

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
#yelpRating = yelpData[0]['businesses'][0]['rating']
#yelpLocName =  yelpData[0]['businesses'][0]['name']
#yelpDistance =  yelpData[0]['businesses'][0]['distance']
#yelpNoReviews =  yelpData[0]['businesses'][0]['review_count']
#yelpCatName = yelpData[0]['businesses'][0]['category'][0]['alias']
```

```python
#creating empty lists 
#yelpRating = []
#yelpLocName =  []
#yelpDistance = []
#yelpNoReviews =  []
#yelpCatName = []
yelpLat = []
yelpLon = []
```

```python
#grabbing rating, locName, distance, noOfReviews 
for i in yelpData:  
    for j in i['businesses']:
        yelpRating.append(j['rating'])
        yelpLocName.append(j['name'])
        yelpDistance.append(j['distance'])
        yelpNoReviews.append(j['review_count'])
```

```python
#grabbing category name 
for i in yelpData:
    for j in i['businesses']:
        yelpCatName.append(j['categories'][0]['alias'])  
```

```python
#index = 0
for i in yelpData:
    j = len(i['businesses'])
    while j != 0:
        yelpLat.append(i['region']['center']['latitude'])
        yelpLon.append(i['region']['center']['longitude'])
        j-=1
```

```python
a = [1,2,3]
b = []
#if you want to add on item 3 times for example:
for i in range(len(a)):
    j = 3
    while j != 0:
        b.append(a[i])
        j-=1
#now b = [1,1,1,2,2,2,3,3,3]
```
